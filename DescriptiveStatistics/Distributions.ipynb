```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Generate a synthetic dataset with 100,000 samples drawn from a normal distribution
# with mean (loc) 0 and standard deviation (scale) 1
data_normal = stats.norm.rvs(size=100000, loc=0, scale=1)
data_normal
```

```python
# Visualize the distribution of the synthetic dataset 'data_normal' using Seaborn
# The 'displot' function creates a histogram with 100 bins by default
sns.displot(data_normal, bins=100)
```

```python
# Generate a synthetic dataset with 100,000 samples drawn from a uniform distribution
# with lower bound (loc) 0 and upper bound (scale) 10
data_uniform = stats.uniform.rvs(size=100000, loc=0, scale=10)
data_uniform
```

```python
# Visualize the distribution of the synthetic dataset 'data_uniform' using Seaborn
# The 'displot' function creates a histogram with 100 bins by default
sns.displot(data_uniform, bins=100)
```

```python
# Generate a synthetic dataset with 1000 samples drawn from a Bernoulli distribution
# with probability of success (p) 0.2
data_bernouli = stats.bernoulli.rvs(size=1000, p=0.2)
data_bernouli
```

```python
# Simulate hospital outcomes over a year for 1000 patients.
icu = 0.3
surgery = 0.2
el = 0.5
data_multinomial = np.random.multinomial(size=365, n=1000, pvals=[icu, surgery, el])
data_multinomial
```

```python
# Visualize the distribution of the synthetic dataset 'data_multinomial' using Seaborn
# The 'displot' function creates a histogram with 100 bins by default
sns.displot(data_multinomial, bins=100)
```
